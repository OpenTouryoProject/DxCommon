#### 算術計算

##### 四則演算
２つのNumPy行列を算術計算で処理する。

###### 要素ごとの足し算

```python
# 行列数が同じ
x=np.array([[1, 2], [3, 4]])
y=np.array([[3, 0], [0, 6]])
x+y
```

```python
# 行列数が異なる（ブロード・キャスト）
x=np.array([[1, 2], [3, 4]])
y=np.array([10,20])
x+y
```

###### 要素ごとの掛け算
※ 行列の積ではないので注意。

```python
# 行列数が同じ
x=np.array([[1, 2], [3, 4]])
y=np.array([[3, 0], [0, 6]])
x*y
```

```python
# 行列数が異なる（ブロード・キャスト）
x=np.array([[1, 2], [3, 4]])
y=np.array([10,20])
x*y
```

##### その他の計算
[配列（ベクトル）の "算術計算 > その他の計算"](#その他の計算) と同じように計算できなかった。  
ベクトルを行列に変換して計算してみると良い（結果がエラーになる）。

#### 統計計算
[配列（ベクトル）の統計計算](#統計計算)と同じように計算できる。  
ベクトルを行列に変換して計算してみると良い（結果が同じになる）。

##### 配列

```python
x=np.random.randn(12)
print(x)
print(np.sign(x))
print(x.mean())
print(x.std())
```

##### 行列

```python
x=np.reshape(x,[4,3]) # ベクトルを行列に変換（後述
print(x)
print(np.sign(x))
print(x.mean())
print(x.std())
```

## ベクトルの積算（内積）

```python
a=a.reshape(1, -1)
b=b.reshape(1, -1).T # 転置行列に
```

```python
print("a")
print(a)
print("b")
print(b)
```

#### 演算子

```python
a@b
```

#### メソッド

```python
np.dot(a,b)
```

### 交換法則が成立
行列では、交換法則の不成立になるが、  
ベクトルの場合は成立する（ただし転置が必要）。

#### 演算子

```python
a@b
```

```python
b.T@a.T
```

#### メソッド

```python
np.dot(a,b)
```

```python
np.dot(b.T,a.T)
```
