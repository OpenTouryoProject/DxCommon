## N次配列操作
- ２次配列は行列とも呼ばれる。
- ３次元以上の配列を、テンソル/多次元配列と呼ぶ。

### 生成

#### 通常

```python
a=np.array([[1,2],[3,4],[5,6]])
```

#### 型指定

```python
a=np.array([[1,2],[3,4],[5,6]], dtype=np.int32)
```

#### 表示

##### 生

```python
a
```

##### print

```python
print(a)
```

##### その他の関数

###### ndim (行列の次元は１次元

```python
np.ndim(a)
```

###### shape (次元の形状

```python
a.shape
```

###### shape[n] (次元の要素数

```python
a.shape[0]
```

###### dtype (列要素のデータ型

```python
a.dtype
```
