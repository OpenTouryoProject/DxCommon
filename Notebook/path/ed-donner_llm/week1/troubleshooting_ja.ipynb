# ステップ3

次に、APIキーが `.env`ファイルで正しいセットアップであり、dotenvパッケージを使用して利用可能であることを確認しましょう。
次のセルを実行してみてください。

```python
# これにより、APIキーを出力に印刷する必要があります - 印刷される指示に従ってください

import os
from dotenv import load_dotenv
load_dotenv(override=True)

api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    print("APIキーが見つかりませんでした - カーネルメニューを試してください>>すべてのセルのカーネルとクリア出力を再起動してください")
elif not api_key.startswith("sk-proj-"):
    print(f"An API key was found, but it starts with {api_key[:8]} rather than sk-proj- please double check this is as expected.")
elif api_key.strip() != api_key:
    print("APIキーが見つかりましたが、開始時または最後にスペースまたはタブ文字があるように見えます - それらを削除してください")
else:
    print("APIキーが見つかり、これまでのところ見栄えがします！")

if api_key:
    problematic_unicode_chars = ['\u2013', '\u2014', '\u201c', '\u201d', '\u2026', '\u2018', '\u2019']
    forbidden_chars = ["'", " ", "\n", "\r", '"']
    
    if not all(32 <= ord(char) <= 126 for char in api_key):
        print("潜在的な問題：キーに誤って印刷できないキャラクターが含まれている可能性がありますか？")
    elif any(char in api_key for char in problematic_unicode_chars):
        print("潜在的な問題：キーに長いハイフンや巻き毛の引用のような特殊文字があるかもしれません - ワードプロセッサでコピーしましたか？")
    elif any(char in api_key for char in forbidden_chars):
        print("潜在的な問題：キーに引用符、スペース、または空の行がありますか？")
    else:
        print("APIキーには有効な文字が含まれています")
    
print(f"\nHere is the key --> {api_key} <--")
print()
print("このキーが良く見える場合は、編集メニュー>>セル出力をクリアして、ここにキーが表示されなくなるようにしてください！")
```
